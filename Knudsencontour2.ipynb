```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

This is the code to plot the Knudsen number. The following equation is of the form: 



$K_n = \frac{c}{\rho \cdot R}$

$where \quad c = \frac{\lambda}{2}$

```python
c = 100**3 * 10**-5 
print(c)
```

```python
#Constants defined in the equation 
#Lambda
mean_free_path = (1.86*10**-6)

#Constant c
constant = mean_free_path/2 
r_constant = 10**-5 * 100**3 * 10**-3

C = constant*r_constant

Radius = (10**-9, 10**-3, 1000)
rho = (10**-17, 1, 1000)

Rad_log = np.logspace(-9,-3,1000)

rho_log = np.logspace(-17,1,1000)



#Converting Radius and rho into a total 2D array 
Rad, Rho = np.meshgrid(Rad_log,rho_log)


Knudsen = (C)/((Rad) * (Rho))


#Contour_plot_1 = plt.contourf(Rad,Rho, Knudsen,levels = [1], cmap = 'magma')
Contour_plot = plt.contour(
    Rad,Rho, Knudsen, 
    levels = np.logspace(-5,10,100), 
    #norm arguement tells you how the colour must be mapped 
    norm = LogNorm(), 
    cmap = 'viridis')
plt.colorbar(label = 'Kn')


Contour_plot2 = plt.contour(
    Rad,Rho, Knudsen, 
    levels = [1], 
    #norm arguement tells you how the colour must be mapped 
    norm = LogNorm(), 
    cmap = 'cividis')


plt.xlabel('Radius ($m$)')
plt.ylabel('Density ($\\frac{kg}{m^3}$)')
plt.xscale('log')
plt.yscale('log')





plt.title('Contour Plot for Knudsen number')

plt.grid(which= 'both', linewidth = 0.3)


plt.show()

















```
