```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#Code for Knudsen number 
```

```python


mean_free_path = 1.86*10**-6


constant = mean_free_path/2 



Radius = np.linspace(10**-9, 10**-3, 100000)
rho = np.linspace(10**-1, 1, 1000000) 


Rad_grid,Rh_grid = np.meshgrid(Radius, rho)





Knudsen_number = constant/(Rad_grid*Rh_grid)
```

```python

fig,ax = plt.subplots(figsize = (10,10),facecolor='black')
ax.set_facecolor('black')


contour = ax.contour(Rad_grid,Rh_grid, Knudsen_number, levels = [0,0.01,0.1,1,10], cmap = 'spring')

plt.clabel(contour, inline = True, fontsize = 9)
plt.colorbar(contour)

ax.tick_params(colors = 'white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.title.set_color('white')
leg = ax.legend()
plt.setp(leg.get_texts(), color='white')   # make text white
leg.get_frame().set_edgecolor('white')     # optional: white border
leg.get_frame().set_facecolor('none')      # optional: transparent background

plt.xlabel('Radius')
plt.xlim(0,0.0004)

plt.ylabel('Density')
plt.grid()
plt.title("Contour plot")



plt.show()
```
