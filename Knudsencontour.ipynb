```python
import matplotlib.pyplot as plt
import numpy as np

# Example data (replace Reynolds with your actual array)
mean_free_path = 1.86e-6
constant = mean_free_path / 2
Radius = np.linspace(1e-9, 1e-3, 1000)
rho = np.linspace(1e-1, 1, 1000)
Rad_grid, Rh_grid = np.meshgrid(Radius, rho)

Knudsen_number = constant / (Rad_grid * Rh_grid)
Reynolds = (Rad_grid * Rh_grid) / constant  # dummy example just for plotting




# ---- Plotting ----
fig, ax = plt.subplots(figsize=(10, 10), facecolor='black')
ax.set_facecolor('black')

# Contours for both variables
contour1 = ax.contour(Rad_grid, Rh_grid, Knudsen_number, 
                      levels=[0, 1,10], 
                      cmap='spring', linewidths=1.2)

contour2 = ax.contour(Rad_grid, Rh_grid, Reynolds, 
                      levels=[0, 1,10], 
                      cmap='winter', linewidths=1.2, linestyles='dashed')

# Add labels directly on contours
plt.clabel(contour1, inline=True, fontsize=9, fmt='Kn=%.2f')
plt.clabel(contour2, inline=True, fontsize=9, fmt='Re=%.2f')

# Create custom legend handles
handles = [
    plt.Line2D([], [], color='magenta', label='Knudsen number (solid)'),
    plt.Line2D([], [], color='cyan', linestyle='--', label='Reynolds number (dashed)')
]
leg = ax.legend(handles=handles, loc='upper right', facecolor='none', edgecolor='white', fontsize=10)

# Style adjustments
ax.tick_params(colors='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
ax.title.set_color('white')
plt.setp(leg.get_texts(), color='white')

plt.xlabel('Radius')
plt.ylabel('Density')
plt.grid(True, color='gray', alpha=0.3)
plt.title("Contour Plot: Knudsen vs Reynolds Numbers")


plt.xlim(0,0.0002)
plt.show()
```
